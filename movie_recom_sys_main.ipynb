```python
movies=movies.merge(credits,on='title')
```

```python
''' id
overview
cast 
crew
title
keywords
genres'''
```

```python
movies=movies[['movie_id','title','overview', 'genres','keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
import ast
```

```python
def convert3(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head()
```

```python
def director_name(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director' :
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(director_name) 
movies.sample(7)
```

```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['crew']=movies['crew'].apply(collapse)
movies['cast']=movies['cast'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(3)
```

```python
new = movies.drop(columns=['overview','cast','crew','keywords','genres'])
```

```python
new.head()
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
```

```python
new.head(2)
```

```python
#vectorization
from sklearn.feature_extraction.text import CountVectorizer
CV=CountVectorizer(max_features=5000, stop_words="english")
```

```python
vector= CV.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title']=='Avatar'].index[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse= True, key= lambda x:x[1])
```

```python
#main function 
def recommend(movie):
    index=new[new['title']==movie].index[0]
    movie_list= sorted(list(enumerate(similarity[index])), reverse= True, key= lambda x:x[1])[1:6]
    for i in movie_list:
        print(new.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```
